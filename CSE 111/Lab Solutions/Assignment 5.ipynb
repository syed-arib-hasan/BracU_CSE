```python
#task1
import math
class Circle:
  def __init__(self,r):
    self.__radius=r

  def area(self):
    return (math.pi*pow(self.__radius,2))

  def getRadius(self):
    return self.__radius

  def setRadius(self,r):
    self.__radius=r

c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" , c1.area())


c2 = Circle(5)
print("Second circle radius:" , c2.getRadius())
print("Second circle area:" , c2.area())
```

```python
#task 2
class Triangle:
  def __init__(self,b,h):
    self.__b=b
    self.__h=h

  def getBase(self):
    return self.__b
  def getHeight(self):
    return self.__h
  def setBase(self,new):
    self.__b=new
  def setHeight(self,new):
    self.__h=new
  
  def area(self):
    return .5*self.__b*self.__h

t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())


t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
```

```python
#task 6
class Train:
  def __init__(self,*n):
    self.tname=n[0]
    self.start=n[1]
    self.des=n[-1]
    self.troute=n[1:]
    self.pas=[]
    self.obj=[]
    print(f"Welcome aboard on {self.tname}")
    print(f"Start: {self.start}")
    print(f"Destination: {self.des}")

  def addPassenger(self,*a):
    for i in a:
      self.pas.append(i.name)
      self.obj.append(i)
      print(i.name,"welcome aboard")

  def allPassengerDetails(self):
    for i in self.obj:
      if i.start==None and i.stop==None:
        i.start=self.start
        i.stop=self.des
      elif i.start!=None and i.stop==None:
        i.stop=self.des
      print(f"Name: {i.name}, Start: {i.start},Destination: {i.stop},Fair: ${(self.troute.index(i.stop)-self.troute.index(i.start))*100}")


class Passenger:
  def __init__(self,n,s=None,d=None):
    self.name=n
    self.start=s
    self.stop=d
    

t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()

```

```python
#task 5
class Student:
  def __init__(self,name,id,dept):
    self.name=name
    self.id=id
    self.dept=dept
    self.email=None
    self.password=None
    self.log=None
    print("Student object is created")
class Usis:
  def __init__(self):
    print("USIS is ready to use!")

  def login(self,a):
    if a.email==None or a.password==None:
      print("Email and password need to be set")
    else:
      print("Login successful!")
      a.log=True

  def advising(self,*a):
    if a[0].log==None:
      print("Please login to advise courses!")
    elif len(a)==1:
      print("You haven't selected any courses.")
    elif len(a)>=5:
      print("You need special approval to take more than 3 courses.")
    else:
      print("Advising successful!")
      self.course=list(a)[1:]
    
  def individualDetails(self,a):
    l=list(self.course)
    return f"Name: {a.name} \nID: {a.id} \nDepartment: {a.dept} \nAdvised course: {','.join(l)}"

rakib = Student("Rakib", 12301455, "CSE")
print("1***********************")
usis_obj = Usis()
print("2***********************")
usis_obj.login(rakib)
print("3***********************")
usis_obj.advising(rakib)
print("4***********************")
rakib.email = "rakib@hotmail.com"
rakib.password = "1234"
print("5***********************")
usis_obj.login(rakib)
print("6***********************")
usis_obj.advising(rakib)
print("7***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110", "CSE260")
print("8***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110")
print("9***********************")
print(usis_obj.individualDetails(rakib))





  
```

```python
#task 7
class BracuStudent:
    def __init__(self,name,loc):
        self.name=name
        self.home=loc
        self.pas=False
    
    def show_details(self):
        print (f"Student Name: {self.name} \nLives in {self.home} \nHave Bus Pass? {self.pas}")
    def get_pass(self):
        self.pas=True


class BracuBus:
    def __init__(self,loc,cap=2):
        self.loc=loc
        self.cap=cap
        self.pas=[]

    def show_details(self):
        print(f"Bus Route: {self.loc} \nPassengers Count: {len(self.pas)} (Max: {self.cap}) \nPassenger On Board:{self.pas}")

    def board (self,*a):
        if len(a)==0:
            print("No passenger!")
        else:           
            for i in a:
              count=len(self.pas)
              if count==self.cap:
                print("Bus is full!")
                break
              elif i.home!=self.loc:
                  print("You got on wrong bus")
              elif i.pas==False:
                  print("You don't have bus pass!")
              else:
                  self.pas.append(i.name)
                  print(i.name,"boarded the bus.")

st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```

```python
class Library:
  def __init__(self,n,av={}):
    self.name=n
    self.av=av
    self.br={}
    

  def details(self):
    print(f"{self.name} Library details \nBorrower details: \n{self.br} \nBooks availability: \n{self.av}")


class Reader:
  def __init__(self,a):
    self.reader=a
    self.count=0
    self.d={}


  def borrow(self,*a):
    for i in range(1,len(a)):
      if a[0].av[a[i]]>0 and self.count<5:
        print(a[i],"book is borrowed successfully.")
        a[0].av[a[i]]=a[0].av[a[i]]-1
        self.count+=1
        if a[i] not in self.d:
          self.d[a[i]]=1
        else:
          self.d[a[i]]=self.d[a[i]]+1
      elif a[0].av[a[i]]<=0:
        print(a[i],"books are not available at the moment.")
      elif self.count>=5:
        print("You cannot borrow more than 5 books.")
    if self.reader not in a[0].br:
      a[0].br[self.reader]=self.count
    else:
      a[0].br[self.reader]=a[0].br[self.reader]+1

    

  def readerInfo(self,a=None):
    if a==None:
      print(f"{self.reader}, you have {self.count} book(s) with you.")
      for i in self.d:
        print(f"Books on {i}: {self.d[i]}")
    else:
      print(f"{self.reader}, you have {self.d[a]} {a} books(s) with you")


L1=Library('Dhaka',{'Arts':15,'Fiction':135,'Politics':2,'Science':11,'Poetry':15})
L1.details()
print("1----------------------")
r1=Reader('Aladdin')
r1.borrow(L1,'Arts','Fiction','Fiction','Politics')
print("2----------------------")
r1.borrow(L1,'Politics','Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2=Reader('Jasmine')
r2.borrow(L1,'Politics','Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()
```
