```python
#task 7
class Match:
  def __init__(self,a):
    print("5..4..3..2..1..Play!")
    self.a=a.split("-")
    self.run=0
    self.w=0
    self.o=0

  def add_run(self,r):
    self.run+=r
  def add_wicket(self,w):
    self.w+=1
  def add_over(self,o):
    if o<5:
      self.o+=o
    elif o>=5:
      print(f"Warning! Cannot add {o} over/s (5 over match)")
  def print_scoreboard(self):
    return(f"Batting Team: {self.a[0]} \nBowling Team: {self.a[1]} \nTotal Runs: {self.run} Wickets: {self.w} Overs: {self.o}")


match1 = Match("Rangpur Riders-Cumilla Victorians")
print("=========================")
match1.add_run(4)
match1.add_run(6)
match1.add_over(1)
print(match1.print_scoreboard())
print("=========================")
match1.add_over(5)
print("=========================")
match1.add_wicket(1)
print(match1.print_scoreboard())

```

```python
#taask 9
class Batsman:
  def __init__ (self, *info):
    if len(info)==2:
      self.run=info[0]
      self.ball=info[1]
      self.name=None
      #print(f"Name: New Batsman")
    elif len(info)==3:
      self.name=info[0]
      self.run=info[1]
      self.ball=info[2]
  def printCareerStatistics(self):
    if self.name!=None:
      print(f"Name: {self.name}")
    else:
      print(f"Name: New Batsman")
    print(f"Runs Scored: {self.run} , Balls Faced: {self.ball}")
  
  def battingStrikeRate(self):
    return((self.run/self.ball)*100)

  def setName(self,name):
    if self.name==None:
      self.name=name

b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#task 12
class TaxiLagbe:
  def __init__(self,a,b):
    self.no=a
    self.area=b
    self.pas=()
    self.fare=0
   
  def addPassenger(self,*n):
    for i in n:
      j=i.split("_")
      self.pas+=(j[0],)     
      if len(self.pas)>=5:
        self.pas=self.pas[:-1:1]
        print("Taxi Full! No more passengers can be added.")
        break
      else:
        self.fare+=int(j[1])
        print(f"Dear {j[0]}! Welcome to TaxiLagbe.")
      
  def printDetails(self):
    print(f"Trip info for Taxi number: {self.no}")
    print(f"This taxi can cover only {self.area} area.")
    print(f"Total passemgers: {len(self.pas)}")
    print("Passenger lists:")
    for i in range(len(self.pas)):
      if i==len(self.pas)-1:
        print(self.pas[i],end="")
      else:
        print(self.pas[i],end=",") 
    print()
    print(f"Total collected fare: {self.fare} Taka")


taxi1 = TaxiLagbe ('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200',"arib_200")
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
#task 14
class StudentDatabase:
  def __init__(self,name,id):
    self.name=name
    self.id=id
    self.grades={}

  def calculateGPA(self,lst,year):
    self.year=year
    self.lst=lst
    sum=0
    self.tup=()
    self.dictup={}
    for i in lst:
      i=i.split(":")
      sum+=float(i[1])
      self.tup+=(i[0],)
    self.gpa=str("{:.2f}".format((sum/len(self.tup))))
    self.dictup[self.tup]=self.gpa
    self.grades[year]=self.dictup
    
  def printDetails(self):
    print("Name: ",self.name)
    print("ID:",self.id) 
    for i in self.grades.keys():
      print(f"Courses taken in {i}")
      for j in self.grades[i].keys():
        for item in j:
          print(item)
        print(self.grades[i][j]) 


s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()




```

```python
class RickMote():
  def __init__(self):
    self.pow="OFF"
    self.ch=0
    self.v=3
    self.lst=[0,2,3,6,7,9]

  def power(self):
    if self.pow=="OFF":
      self.pow="ON"
    elif self.pow=="ON":
      self.pow="OFF"

  def changeChannel(self,a=2):
    if self.pow=="ON":
      if a==2:
        self.ch=2
      elif a in self.lst:
        self.ch=a
      else:
        print("TV channel does not exist.")
    else:
      print("Power is turned off. Cannot change volume.")

  def changeVolumeLevel(self,v=1):
    if self.pow=="ON":
      self.v+=v
    else:
      print("Power is turned off. Cannot change volume.")

  def showInfo(self):
    print("ID Cable Box Status:")
    print("Cable Box is:",self.pow)
    if self.pow=="ON":
      print("Channel:",self.ch)
      print("Volume:",self.v)

oTV = RickMote()
oTV.power()
print("1.############################")
oTV.showInfo()
print("2.############################")
oTV.changeChannel()
oTV.changeVolumeLevel()
oTV.showInfo()
print("3.############################")
oTV.power()
oTV.showInfo()
print("4.############################")
oTV.power()
oTV.changeVolumeLevel(4)
oTV.changeChannel(3)
oTV.showInfo()
print("5.############################")
oTV.changeVolumeLevel(-2)
oTV.showInfo()
print("6.############################")
oTV.power()
oTV.changeChannel(9)
oTV.changeVolumeLevel(-1)
oTV.showInfo()
print("7.############################")
oTV.power()
oTV.changeChannel(11)
oTV.showInfo()
```

```python
class PizzaMachine():
  def __init__(self,p="Regular",sz=6):
    self.p=p
    self.sz=sz
    self.sp=["Regular", "Hot","Super Naga"]

  def customizePizza(self,lst="None",sp="Regular"):
    if type(lst)!=list:
      return("No toppings specified! Can't bake pizza.")
    elif sp in self.sp:
      temp=""
      for i in range(len(lst)):
        if i==len(lst)-1:
          temp+=lst[i]
        else:
          temp+=lst[i]+","

      return(f"Your {self.sz}-inch {sp} spicy {self.p} Pizza is ready with {temp} toppings. Enjoy!")
    else:
      return("Sorry! Spice level not allowed. Can't bake pizza.")
pizza1 = PizzaMachine()
order1 = pizza1.customizePizza(["Cheese", "Pepperoni"],"Hot")
print("1################################# ")
print(order1)
print("2================================")
pizza2 = PizzaMachine("Vege")
order2 = pizza2.customizePizza("Super Naga")
print("3#################################")
print(order2)
print("4================================")
pizza3 = PizzaMachine("Chicken Blast",12)
order3 = pizza3.customizePizza(["Mushroom"])
print("5#################################")
print(order3)
print("6================================")
pizza4 = PizzaMachine("Beef Bonanza",16)
order4 = pizza4.customizePizza(["Cheese","Beef kalabhuna"],"Mild")
print("7#################################")
print(order4)
print("8================================")
```

```python
class InstaUser():
  def __init__(self,name,a=None,b=None):
    if a==None:
      print("Birth year needed for profile creation!")
      self.year="Unknown"
      self.status=False
    else:
      self.year=a
      self.status=True
    if b==None:
      self.country="USA"
    else:
      self.country=b
    self.frnd=[]
    self.name=name
    if self.status==True:
      print("Profile Created Successfully!")

  def showInfo(self):
    return(f"Name: {self.name} \nCountry: {self.country} \nBirth Year: {self.year} \nProfile Status: {self.status} \nTotal Friends: {len(self.frnd)}")

  def addFriends(self,lst):
    for i in lst:
      self.frnd.append(i)

  def showStory(self,a):
    if a in self.frnd:
      print(f"{a}'s story is loading...")
    else: 
      print (f"You are not friends with {a}!")


user1 = InstaUser('John')
print('1.==============================')
print(user1.showInfo())
print('2.==============================')
user2 = InstaUser('Shoaib', 1992, 'Bangladesh')
print('3.==============================')
user2.addFriends(['Imtiaz', 'Tanvir', 'Noman'])
print('3.==============================')
user2.showStory('Imtiaz')
print('3.==============================')
user2.showStory('Borshon')
print('3.==============================')
print(user2.showInfo())
```
